```python
# 1. DEFINE THE RULES
# We'll flag meetings with more than 10 people as "Low Efficiency"
df['Efficiency'] = df['Attendees'].apply(lambda x: 'High' if x <= 10 else 'Low (Musk Violation)')

# 2. CALCULATE THE "WASTE"
# Let's assume every Low Efficiency meeting wastes 45 minutes of a CEO's time
minutes_per_bad_meeting = 45
bad_meetings_count = len(df[df['Efficiency'] == 'Low (Musk Violation)'])
total_hours_saved = (bad_meetings_count * minutes_per_bad_meeting) / 60

# 3. PRINT THE EXECUTIVE SUMMARY
print("--- EXECUTIVE AI SUITE: STRATEGIC REPORT ---")
print(f"Total Meetings Analyzed: {len(df)}")
print(f"Inefficient Meetings Found: {bad_meetings_count}")
print(f"POTENTIAL TIME RECOVERED: {total_hours_saved} HOURS")
print("-" * 40)

# 4. SHOW ONLY THE MEETINGS TO CANCEL
print("\n[ACTION REQUIRED] Recommended Cancellations:")
display(df[df['Efficiency'] == 'Low (Musk Violation)'][['Meeting Title', 'Attendees', 'KPI']])
```

```python
import matplotlib.pyplot as plt

# 1. PREPARE THE VISUAL DATA
# Total meetings: 10 (High Efficiency: 7 | Low Efficiency: 3)
efficiency_labels = ['Efficient (High-Value)', 'Inefficient (Musk Violation)']
meeting_counts = [7, 3] # These are the counts from your analysis
colors = ['#2ecc71', '#e74c3c'] # Professional Green and Red

# This "explodes" the inefficient slice to make it stand out
explode = (0, 0.1) 

# 2. CREATE THE HIGH-TECH CHART
plt.figure(figsize=(10, 7)) # Set a good size for the chart

# Start creating the pie
plt.pie(meeting_counts, explode=explode, labels=efficiency_labels, 
        colors=colors, autopct='%1.1f%%', shadow=True, 
        startangle=140, textprops={'fontsize': 12})

# 3. ADD THE EXECUTIVE TITLE
plt.title('EXECUTIVE TIME ALLOCATION: EFFICIENCY REPORT', 
          fontsize=16, fontweight='bold', pad=20)

# Make the pie a perfect circle
plt.axis('equal') 

# 4. SHOW THE CHART
plt.show()
```
